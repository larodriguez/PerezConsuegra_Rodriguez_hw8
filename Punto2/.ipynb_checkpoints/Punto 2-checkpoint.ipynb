## Aqui implemento el codigo para obtener el hessiano. Para esto creo una matriz T (Tensor) y alli voy a almacenar los datos. Siguiendo el metodo propuesto por Forero en:http://nbviewer.ipython.org/github/forero/ComputationalMethods/blob/master/notes/Tensors.ipynb

```python
h = 1
#definimos valores aproximados de las filas entre las cuales estan los bordes de la vesicula que queremos encontrar

Value1 = zeros((col))
Value2 = zeros((col))
Value3 = zeros((col))
Value4 = zeros((col))
Value5 = zeros((col))
Value6 = zeros((col))
T=ones((2,2))
for x in range(1,fil-1):
    
    for y in range(1,col-1):
        
        for i in range(2):
            
            for j in range(2):
        
                if(i==0 and j==0):
                
                    T[0,0] = ( normal1[x+h,y] + normal1[x-h,y] -  2*normal1[x,y] )/ h**2
        
                elif(i==1 and j==1):
        
                    T[1,1]=  normal1[x,y+h] + normal1[x,y+h] -  2*normal1[x,y] / (h**2)
                                                                    
                elif(i==1 and j==0):
        
            
                    T[0,1]= (normal1[x+h][y+h] + normal1[x-h][y-h] + 2*normal1[x][y] - normal1[x+h][y] -normal1[x-h][y] -normal1[x][y+h]-normal1[x][y-h] )/(2*(h**2))
       
                elif(i==0 and j==1):
                    
                    T[1,0]= T[0,1]
                    
        eigval,eigvec = eig(T)
        if(x==150):
            
            Value1[y] = eigval[1]
            Value2[y] = eigval[0]
            Value3[y] = eigval[0]*eigval[1]
            Value4[y] = abs(eigval[0]*eigval[1])
            Value5[y] = np.linalg.norm(eigvec[0])
            Value6[y] = np.linalg.norm(eigvec[1])

                  
```

## En el punto anterior cree un archivo en el cual guarde todas las coordenadas x, y d elo spuntos que cumplian la condicion abs(eigval >=1600), es decir, los puntos sobre el borde. Aqui los ploteo...

```python
filename = 'coordenadas.txt'
file = open(filename, 'r') #Abrir el archivo para escribir
a,b= loadtxt(file, unpack=True)

scatter(a,b)
```

```python
### Taken from:http://docs.scipy.org/doc/scipy/reference/tutorial/optimize.html#least-square-fitting-leastsq

from numpy import arange, sin, pi, random, array


angulo_theta = arange(0, 2*pi, 2*pi/len(a))
print angulo_theta
r1, r0, phi,m = 1000, 1.0 / 3e-2, pi / 2, pi/2

y_true = r0 + r1*sin(m*angulo_theta+ phi)

print m


y_meas = a
```

```python
def residuals(p, y, angulo_theta):
    r1, r0, phi,m = p
    err = y - r0 + r1*sin(m*angulo_theta + phi)
    return err

def peval(angulo_theta, p):
    return p[1] + p[0]*sin(p[3]*angulo_theta+ p[2])


p0 = [10, 1 / 2.3e-20, pi / 3,3*pi/2]
print(array(p0))

from scipy.optimize import leastsq
plsq = leastsq(residuals, p0, args=(y_meas, angulo_theta))

#
#print array([r1, r0, phi,m ])
print(plsq[0])
plt.plot(angulo_theta, peval(angulo_theta, plsq[0]),angulo_theta,y_meas,'o',angulo_theta,y_true)
plt.title('Least-squares fit to noisy data')
plt.legend(['Fit', 'Noisy', 'True'])
plt.show()
r1, r0, phi,m = plsq[0]
y = r0 + r1*sin(m*angulo_theta+ phi)
```
