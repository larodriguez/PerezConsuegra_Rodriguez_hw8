El objetivo es encontrar el centro y los puntos que describen el contorno de las vesículas. Para esto vamos a tomar que desde el centro de la vesícula el contorno cumple la siguiente ecuación en coordenadas polares: $$r(\theta) = r_0 + r_1 sin(m\theta + \phi)$$ 

Donde las constantes son parametros libres con m > 0 entero.


Para esto se realiza un programa que utiliza MCMC para encontrar las distribuciones de probabilidad de la posición del centro del circulo y los parametros libres.

```python
#importar librerias
%pylab inline 
import numpy as np
```

Llamamos la imagen original, que se encuentra en el repositorio

```python
#llamamos la primera imagen
imagen1 = pylab.imread("guv_spim0050.tif")
fig = figure(1, figsize=(8,8))
imshow(imagen1[:,:], cmap = cm.Greys_r)
plt.title('$Imagen\ 1\ Original$', fontsize=20)
```

Para el desarrollo del metodo vamos a utilizar la imagen como una matriz, pero para esto asignaremos valores entre 0 y 1 a la matriz de manera que este Normalizada

```python
#creamos una matriz de zeros del mismo tamaño de la imagen
y = imagen1.astype(np.float)
a = np.zeros(np.shape(y))
normal1 = np.zeros_like(a)
f = amax(y)
#asignamos valores de 1 en los bordes de las veciculas (NORMALIZAMOS)
for i in range (327):
    for j in range (322):
            normal1[i,j] = abs(y[i,j]/f)
```

```python
#imprimimos la imagen normalizada
fig = figure(1, figsize=(8,8))
imshow(normal1[:,:], cmap = cm.Greys_r)
plt.title('$Imagen\ 1\ Normalizada$', fontsize=20)
```

Definimos la función "my_model" la cual recime los parametros libres y la variable por parametro y retorna el valor equivalente de r tal como se describe en la ecuación inicial

```python
def my_model(theta, r0, r1, m, phi):
    return r0 + (r1*sin((m*theta)+phi))
```

Definimos la función  "loglikelihood" la cual retorna: $$-x^2$$

con la cual podemos hacer la comparación entre la imagen real y el modelo, la razón por la cual no definimos "likelihood" es que la exponencial se hace 0 y no nos funciona asi para este caso

```python
def loglikelihood(imagen_obs, imagen_model):
    chi_squared = (1.0/2.0)*sum((imagen_obs-imagen_model)**2)
    return -chi_squared
```

Hacemos correr el modelo para una condición inicial que ajustamos mirando la imagen original que tenemos

```python
r0_walk = empty((0))
r1_walk = empty((0))
m_walk = empty((0))
phi_walk = empty((0))
l_walk = empty((0))

x_walk = empty((0))
y_walk = empty((0))

r0_walk = append(r0_walk, 100)
r1_walk = append(r1_walk, 1)
m_walk = append(m_walk, 10)
phi_walk = append(phi_walk, (random.random()*2*pi))

#x0 y y0
x_walk = append(x_walk, 160)
y_walk = append(y_walk, 130)

theta = linspace(0,2*pi,1000)


b = np.zeros(np.shape(imagen1))
imagen_init = np.zeros_like(b)

for i in theta:
    x = int((my_model(i, r0_walk[0], r1_walk[0], m_walk[0], phi_walk[0])*cos(i)) + x_walk[0])
    y = int((my_model(i, r0_walk[0], r1_walk[0], m_walk[0], phi_walk[0])*sin(i)) + y_walk[0])
    
    imagen_init[x,y] = 1
    
l_walk = append(l_walk, loglikelihood(normal1, imagen_init))

print l_walk


fig = figure(1, figsize=(8,8))
imshow(imagen_init[:,:], cmap = cm.Greys_r)
plt.title('$Imagen\ 1\ Modelo\ Inicial$', fontsize=20)
```

Corremos el modelo por un determinado numero n de itereciones 

```python
n_iterations = 5000
for i in range(n_iterations):
    r0_prime = np.random.normal(r0_walk[i], 1)
    r1_prime = np.random.normal(r1_walk[i], 1)
    m_prime = np.random.normal(m_walk[i], 1)
    phi_prime = np.random.normal(phi_walk[i], 0.1)
    
    x_prime = np.random.normal(x_walk[i], 1)
    y_prime = np.random.normal(y_walk[i], 1)
    
    while (r0_prime<0):
        r0_prime = np.random.normal(r0_walk[i], 1)
    while (r1_prime<0):
        r1_prime = np.random.normal(r1_walk[i], 1)
    while (m_prime<1):
        m_prime = np.random.normal(m_walk[i], 1)
    while (phi_prime<0 or phi_prime>2*pi):
        phi_prime = np.random.normal(phi_walk[i], 0.1)
    while (x_prime<0):
        x_prime = np.random.normal(x_walk[i], 1)
    while (y_prime<0):
        y_prime = np.random.normal(y_walk[i], 1)
    
    theta = linspace(0,2*pi,1000)


    d = np.zeros(np.shape(imagen1))
    imagen_init = np.zeros_like(d)

    x = int_((my_model(theta, r0_walk[i], r1_walk[i], m_walk[i], phi_walk[i])*cos(theta)) + x_walk[i])
    y = int_((my_model(theta, r0_walk[i], r1_walk[i], m_walk[i], phi_walk[i])*sin(theta)) + y_walk[i])
    
    imagen_init[x,y] = 1
    
    e = np.zeros(np.shape(imagen1))
    imagen_prime = np.zeros_like(e)

   
    xx = int_((my_model(theta, r0_prime, r1_prime, m_prime, phi_prime)*cos(theta)) + x_prime)
    yy = int_((my_model(theta, r0_prime, r1_prime, m_prime, phi_prime)*sin(theta)) + y_prime)
    
    imagen_prime[xx,yy] = 1
    
    
    l_prime = loglikelihood(normal1, imagen_prime)
    l_init = loglikelihood(normal1, imagen_init)
    
    alpha = exp(l_prime-l_init)
    
    if(alpha>=1.0):
        r0_walk  = append(r0_walk,r0_prime)
        r1_walk  = append(r1_walk,r1_prime)
        m_walk  = append(m_walk,m_prime)
        phi_walk  = append(phi_walk,phi_prime)
        x_walk  = append(x_walk,x_prime)
        y_walk  = append(y_walk,y_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            r0_walk  = append(r0_walk,r0_prime)
            r1_walk  = append(r1_walk,r1_prime)
            m_walk  = append(m_walk,m_prime)
            phi_walk  = append(phi_walk,phi_prime)
            x_walk  = append(x_walk,x_prime)
            y_walk  = append(y_walk,y_prime)
            l_walk = append(l_walk, l_prime)
        else:
            r0_walk  = append(r0_walk,r0_walk[i])
            r1_walk  = append(r1_walk,r1_walk[i])
            m_walk  = append(m_walk,m_walk[i])
            phi_walk  = append(phi_walk,phi_walk[i])
            x_walk  = append(x_walk,x_walk[i])
            y_walk  = append(y_walk,y_walk[i])
            l_walk = append(l_walk, l_init)
```
